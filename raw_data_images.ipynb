```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def show_data(tif_image_path):
    image = Image.open(tif_image_path)

    # Convert the image to a NumPy array for plotting
    image_array = np.array(image)

    # Plot the image
    plt.figure(figsize=(8, 8))
    plt.imshow(image_array, cmap="gray")
    plt.colorbar(label="Pixel Intensity")
    plt.title(tif_image_path)
    plt.axis()
    plt.show()
```

```python
tif_image_path = "200kV_80kX_OneView_0044.tif"
show_data(tif_image_path)
```

```python
tif_image_path = "200kV_100kX_OneView_0040.tif"
show_data(tif_image_path)
```
